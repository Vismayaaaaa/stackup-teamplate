```python
from transformers import pipeline

# Load the pre-trained GPT-3 model (or a similar variant available on Hugging Face)
generator = pipeline("text-generation", model="gpt2")  # GPT-3 is not available for free, so we use GPT-2 as a placeholder

# Sample loan application data
loan_application = {
    "credit_score": 720,
    "income": 50000,
    "loan_amount": 15000,
    "employment_status": "Full-time"
}

# Simulated decision by the AI model
decision = "approve"

# Construct the prompt for generating the explanation
prompt = (
    f"The AI model has decided to {decision} the loan application. "
    "Here are the reasons for this decision: "
    f"1. The applicant has a credit score of {loan_application['credit_score']}, which is considered good. "
    f"2. The applicant's income of ${loan_application['income']} is sufficient to cover the loan amount of ${loan_application['loan_amount']}. "
    f"3. The applicant is employed full-time, indicating a stable source of income."
)

# Generate the explanation
explanation = generator(prompt, max_length=150, num_return_sequences=1)

# Display the explanation
print(explanation[0]['generated_text'])
```
